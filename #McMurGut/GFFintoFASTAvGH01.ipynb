# Script GFF into FASTA
Version GitHub 01

Decompress all .gff.gz to .gff files in the current working folder  
Transform all .GFF into .FNA  
GZIP all .fna-files in the current folder into .fna.gz

```python
#Decompress all .gff.gz to .gff files in the current working folder  
import os
import gzip
import shutil

# Use the current working directory
folder_path = os.getcwd()

for filename in os.listdir(folder_path):
    if filename.endswith(".gff.gz"):
        gz_file_path = os.path.join(folder_path, filename)
        with gzip.open(gz_file_path, 'rb') as f_in:
            unzipped_file_path = os.path.join(folder_path, filename[:-3])  # Removing '.gz' from filename
            with open(unzipped_file_path, 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)
        print(f"Decompressed {filename} to {filename[:-3]}")
```

```python
####################################
# Transform all .GFF (ALL in current folder!!!) into .FNA
#
# Some sequencing output is given in .gff format, containing mainly genome annotation,
# but also sometimes the full genome at the end of the file.
# This tool extracts the genome sequence from a XXX.gff file in FASTA format and
# saves it as XXX.fna. Everything after ##FASTA is copied to the new file.

####################################
import sys
import os
import pandas as pd
import glob


files = glob.glob('*.gff')
for file in files:
    print(file)
    filepath_gff = str(file)
    togg = False
    print("toggle:", togg)

    pre_ou = filepath_gff
    print("pre_ou:", pre_ou)

    # Get the file's name, and split .gff away from basename to get the file's name only.
    ou = os.path.basename(pre_ou).split('.gff', 1)[0]
    print('ou:', ou)

    # Load the gff file
    fi = open(filepath_gff)
    print("fi:", fi)

    # Create a new file and call it basename + .fna ending in write mode
    w = open(ou+".fna",'w')
    print("w", w)

    for line in fi:
        if "##FASTA" in line:
            togg = True
        elif togg:
            w.write(line)
        else:
            continue

    print("last line:", line)
    print("togg (should be true):", togg)

    fi.close()
    w.close()

    print("File was created as "+ou+".fna")
    print("---- Endloop of "+filepath_gff+" ----")
print("Program finished.")
    
```

```python
####################################
# GZIP all .fna-files in the current folder into .fna.gz
#
# Some sequencing output is given in .gff format, containing mainly genome annotation,
# but also sometimes the full genome at the end of the file.
# This tool extracts the genome sequence from a XXX.gff file in FASTA format and
# saves it as XXX.fna. Everything after ##FASTA is copied to the new file.
#
####################################import os
import gzip
import shutil

folder_path = os.getcwd()

for filename in os.listdir(folder_path):
    if filename.endswith(".fna"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'rb') as f_in:
            gz_file_path = f"{file_path}.gz"
            with gzip.open(gz_file_path, 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)
        print(f"Gzipped {filename} to {filename}.gz")
    print("-------------------")
print("Finished! G-zipping of all .fna files to .fna.gz completed in", folder_path)
        
```
